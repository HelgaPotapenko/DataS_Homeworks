### –í—Å–µ –∑–∞–¥–∞–Ω–∏—è –Ω—É–∂–Ω–æ –≤—ã–ø–æ–ª–Ω–∏—Ç—å —Ç–æ–ª—å–∫–æ —Å –ø–æ–º–æ—â—å—é sql

1. –í—ã–±–µ—Ä–∏—Ç–µ –∑–∞–∫–∞–∑—á–∏–∫–æ–≤ –∏–∑ –ì–µ—Ä–º–∞–Ω–∏–∏, –§—Ä–∞–Ω—Ü–∏–∏ –∏ –ú–∞–¥—Ä–∏–¥–∞, –≤—ã–≤–µ–¥–∏—Ç–µ –∏—Ö –Ω–∞–∑–≤–∞–Ω–∏–µ, —Å—Ç—Ä–∞–Ω—É –∏ –∞–¥—Ä–µ—Å.

```python
tasks+=1
pd.read_sql(
    '''
    SELECT CustomerName, Country, Address 
    FROM Customers 
    WHERE upper(City) = 'MADRID' or upper(Country) in ('FRANCE','GERMANY')
    ''',
    con,
)
```

2. –í—ã–±–µ—Ä–∏—Ç–µ —Ç–æ–ø 3 —Å—Ç—Ä–∞–Ω—ã –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –∑–∞–∫–∞–∑—á–∏–∫–æ–≤, –≤—ã–≤–µ–¥–∏—Ç–µ –∏—Ö –Ω–∞–∑–≤–∞–Ω–∏—è –∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–∞–ø–∏—Å–µ–π.

```python
tasks+=1
pd.read_sql(
    '''
    SELECT Country, count(*) as CustomersCount
    FROM Customers 
    group by Country
    order by count(*) desc
    limit 3
    ''',
    con,
)
```

3. –í—ã–±–µ—Ä–∏—Ç–µ –ø–µ—Ä–µ–≤–æ–∑—á–∏–∫–∞, –∫–æ—Ç–æ—Ä—ã–π –æ—Ç–ø—Ä–∞–≤–∏–ª 10-–π –ø–æ –≤—Ä–µ–º–µ–Ω–∏ –∑–∞–∫–∞–∑, –≤—ã–≤–µ–¥–∏—Ç–µ –µ–≥–æ –Ω–∞–∑–≤–∞–Ω–∏–µ, –∏ –¥–∞—Ç—É –æ—Ç–ø—Ä–∞–≤–ª–µ–Ω–∏—è.

```python
tasks+=1
pd.read_sql(
    '''
    SELECT ShipperName, OrderDate 
    FROM Orders o
        left join Shippers s on s.ShipperID=o.ShipperID
    order by OrderDate
    limit 9, 1
    ''',
    con,
)
```

4. –í—ã–±–µ—Ä–∏—Ç–µ —Å–∞–º—ã–π –¥–æ—Ä–æ–≥–æ–π –∑–∞–∫–∞–∑, –≤—ã–≤–µ–¥–∏—Ç–µ —Å–ø–∏—Å–æ–∫ —Ç–æ–≤–∞—Ä–æ–≤ —Å –∏—Ö —Ü–µ–Ω–∞–º–∏.

```python
tasks+=1
pd.read_sql(
    '''
    with OrderIDs as (
       SELECT OrderId, sum(Quantity*Price) as Cost
       FROM OrderDetails od 
           left join Products p on p.ProductID=od.ProductID
       group by OrderID
       order by Cost desc
       limit 1
       )
    select o.OrderId, ProductName, Quantity, Price
    from OrderIDs o 
        join OrderDetails od on od.OrderID=o.OrderID
        left join Products p on p.ProductID=od.ProductID
    ''',
    con,
)
```

5. –ö–∞–∫–æ–π —Ç–æ–≤–∞—Ä –±–æ–ª—å—à–µ –≤—Å–µ–≥–æ –∑–∞–∫–∞–∑—ã–≤–∞–ª–∏ –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –µ–¥–∏–Ω–∏—Ü —Ç–æ–≤–∞—Ä–∞, –≤—ã–≤–µ–¥–∏—Ç–µ –µ–≥–æ –Ω–∞–∑–≤–∞–Ω–∏–µ –∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –µ–¥–∏–Ω–∏—Ü –≤ –∫–∞–∂–¥–æ–º –∏–∑ –∑–∞–∫–∞–∑–æ–≤.

```python
tasks+=1
pd.read_sql(
    '''
    with ProductIDs as (
            select p.ProductID,ProductName
            from OrderDetails od
                join Products p on p.ProductID=od.ProductID
            group by p.ProductID, ProductName
            order by sum(Quantity) desc
            limit 1
            )
    select OrderID, ProductName, Quantity
    from ProductIDs p
        join OrderDetails od on od.ProductID=p.ProductID
    ''',
    con,
)
```

6. –í—ã–≤–µ–¥–∏—Ç–µ —Ç–æ–ø 5 –ø–æ—Å—Ç–∞–≤—â–∏–∫–æ–≤ –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –∑–∞–∫–∞–∑–æ–≤, –≤—ã–≤–µ–¥–∏—Ç–µ –∏—Ö –Ω–∞–∑–≤–∞–Ω–∏—è, —Å—Ç—Ä–∞–Ω—É, –∫–æ–Ω—Ç–∞–∫—Ç–Ω–æ–µ –ª–∏—Ü–æ –∏ —Ç–µ–ª–µ—Ñ–æ–Ω.

```python
tasks+=1
pd.read_sql(
    '''
    SELECT SupplierName, Country, ContactName, Phone, count(*) as OrdersCount
    FROM (
        Select distinct OrderID, s.SupplierID, SupplierName, Country, ContactName, Phone
        from Suppliers s
            join Products p on p.SupplierID=s.SupplierID
            join OrderDetails od on od.ProductId=p.ProductId) os
    group by SupplierID, SupplierName, Country, ContactName, Phone
    order by count(*) desc
    limit 5
    ''',
    con,
)
```

7. –ö–∞–∫—É—é –∫–∞—Ç–µ–≥–æ—Ä–∏—é —Ç–æ–≤–∞—Ä–æ–≤ –∑–∞–∫–∞–∑—ã–≤–∞–ª–∏ –±–æ–ª—å—à–µ –≤—Å–µ–≥–æ –ø–æ —Å—Ç–æ–∏–º–æ—Å—Ç–∏ –≤ –ë—Ä–∞–∑–∏–ª–∏–∏, –≤—ã–≤–µ–¥–∏—Ç–µ —Å—Ç—Ä–∞–Ω—É, –Ω–∞–∑–≤–∞–Ω–∏–µ –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ –∏ —Å—É–º–º—É.

```python
tasks+=1
pd.read_sql(
    '''
    select  c.Country, ctg.CategoryName, sum(od.Quantity*p.Price) as Cost
    from Orders o 
        join OrderDetails od on od.OrderID = o.OrderID 
        join Products p on p.ProductId= od.ProductId
        left join Categories ctg on ctg.CategoryID=p.CategoryID
        join Customers c on c.CustomerID=o.CustomerID
    where upper(c.Country)='BRAZIL'
    group by p.CategoryID, ctg.CategoryName, c.Country
    order by Cost desc
    limit 1
    ''',
    con,
)
```

8. –ö–∞–∫–∞—è —Ä–∞–∑–Ω–∏—Ü–∞ –≤ —Å—Ç–æ–∏–º–æ—Å—Ç–∏ –º–µ–∂–¥—É —Å–∞–º—ã–º –¥–æ—Ä–æ–≥–∏–º –∏ —Å–∞–º—ã–º –¥–µ—à–µ–≤—ã–º –∑–∞–∫–∞–∑–æ–º –∏–∑ –°–®–ê.

```python
tasks+=1
pd.read_sql(
    '''
    with OrdersCost as (
               SELECT sum(Quantity*Price) as Cost
           FROM Orders o
               join OrderDetails od on od.OrderId = o.OrderId
               left join Products p on p.ProductID = od.ProductID
               join Customers c on c.CustomerID = o.CustomerID
           where upper(c.Country) = 'USA'    
           group by o.OrderId, c.CustomerID)
       
    select max(Cost) - min(Cost) as CostDifference from OrdersCost 
    ''',
    con,
)
```

9. –í—ã–≤–µ–¥–∏—Ç–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–∞–∫–∞–∑–æ–≤ —É –∫–∞–∂–¥–æ–≥–æ –∏—Ö —Ç—Ä–µ—Ö —Å–∞–º—ã—Ö –º–æ–ª–æ–¥—ã—Ö —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–æ–≤, –∞ —Ç–∞–∫–∂–µ –∏–º—è –∏ —Ñ–∞–º–∏–ª–∏—é –≤–æ –≤—Ç–æ—Ä–æ–π –∫–æ–ª–æ–Ω–∫–µ.

```python
tasks+=1
pd.read_sql(
    '''
    select DATE(substr(BirthDate,7,4)||'-'||substr(BirthDate,4,2)||'-'||substr(BirthDate,1,2)) as BDFormated
         , e.LastName||' '||e.FirstName  as FullName  
         , count(*) as OrdersCount
    from Employees e
        left join Orders o on o.EmployeeID=e.EmployeeID
    group by e.EmployeeID, e.LastName, e.FirstName, BDFormated
    order by BDFormated desc
    limit 3
    ''',
    con,
)
```

10. –°–∫–æ–ª—å–∫–æ –±–∞–Ω–æ–∫ –∫—Ä–∞–±–æ–≤–æ–≥–æ –º—è—Å–∞ –≤—Å–µ–≥–æ –±—ã–ª–æ –∑–∞–∫–∞–∑–∞–Ω–æ.

```python
tasks+=1
pd.read_sql(
    '''
    select sum(od.Quantity * CAST(substr(Unit, 1, instr(Unit, ' '))as integer)) as AllTinsCount
    from Products p
        join OrderDetails od on od.ProductId=p.ProductId
    where upper(ProductName) like '%CRAB_MEAT%'
    ''',
    con,
)
```
